# Общая информация по проекту

###### Заказчик:

Сервис по подбору репетиторов Repetit.ru

###### Описание проекта:

Сервис Repetit.ru работает с большим количеством заявок от клиентов с данными о предмете, желаемой стоимости, возрасте ученика, целью занятий и тд. К сожалению, 7 из 8 не доходят до оплаты, при этом обработка заявки консультантом увеличивает конверсию в оплату на 30%. 
Проблема в том, что консультантов не хватает на все заявки и получается, что чем больше заявок — тем меньше конверсия из заявки в оплату и консультанты тратят время на бесперспективные заявки.

###### Задачи:

Разработать модель, которая по имеющейся информации о клиенте и заявке будет предсказывать вероятность оплаты заявки клиентом. Заказчик хочет понять, какие заявки будут оплачены, а какие нет, чтобы одни обрабатывать вручную консультантами, а другие нет. Оценка качества модели будет производиться с использованием precision и ROC-AUC.

##### Обработаем явные дубликаты.

Заказчик указал, что сложность базы данны состоит в болльшом количестве дубликатов.

"На одну заявку может быть только один “назначенный репетитор”, то есть репетитор, с которым случился обмен контактами. Если клиент хочет ещё с одним репетитором обменяться номерами, то мы создаём дубль заявки и назначаем репетитора на дубль, при этом в дубле создаётся запись original_order_id = номер первой заявки. Иногда консультанты создают новую заявку сами и не указывают, что эта заявка — дубль."

Поэтому в базе есть умышленные дубликаты, которые надо будет обработать специальным образом и явные, появившиеся по ошибке.

```python
print ('Количество явных дубликатов в базе:', orders['id'].duplicated().sum())
orders[orders['id'].duplicated()].head()
```

```python
orders[orders['id'] == 2398558]
```

```python
orders = orders.drop_duplicates()
print("Количество дубликатов после отчистки:", orders['id'].duplicated().sum())
```

id всех заявок должен быть уникальным номером, то, что это намеренный дубликат будет указано в поле original_order_id. Поэтому можем удалить все совпадающие id. Перепроверив вручную мы нашли, что время звявки и остальные признаки точно совпадают.

Создадим новый столбец с бинарной классификацией. К положительному исходу-оплате будут соответствовать статусы 6, 13 (дублирующие друг друга), 14 (частичная оплата), 15 (подлежит оплате). Остальные будут отрицательным исходом.

```python
# группируем по original_order_id и записываем все значения 'status_id' в новый столбец 'second_column_status_id'
grouped_status_ids = orders.groupby('original_order_id')['status_id'].unique()
orders['second_column_status_id'] = orders['original_order_id'].map(grouped_status_ids)

# в заявках без дубликатов дублируем значения из 'status_id'
orders['status_id']=orders['status_id'].astype('float')
orders['second_column_status_id'].fillna(orders['status_id'], inplace=True)

# создаем колонку target 
def payment_result(df):
    status = df['second_column_status_id']
    if isinstance(status, float):
        if status in [6, 13, 14, 5, 15]:
            return 1
        else:
            return 0
    else:
        if any(s in [6, 13, 14, 5, 15] for s in status):
            return 1
        else:
            return 0

orders['target'] = orders.apply(payment_result, axis=1)
orders.drop('second_column_status_id', axis=1, inplace=True)
```

### Функция для обработки purpose

Достанем полезные данные из колонки 'purpose'. В ней пользователи отвечают на дополнительные вопросы по выбранным предметам и оставляют заметки в свободной форме. Запишем данные о категории ученика и изучаемом разделе в отдельные признаки.

```python
nlp = spacy.load("ru_core_news_sm")
```

```python
def nlp_processing(row, keyword):
    text = row['purpose']
    result = [] 
    if text is not None:
        lines=text.split('\n')
        for line in lines:
            line = line.strip()
            if line.startswith(keyword):
                line = line[len(keyword):].strip()
                first_dot = line.find('.')
                sent = line[:first_dot]
                doc = nlp(sent)
                lemmatized_text = ' '.join([token.lemma_ for token in doc])
                result.append(lemmatized_text) 
    return ', '.join(result) if result else None

#orders['nlp_aim'] = orders.apply(lambda row: nlp_processing(row, 'Разделы:'), axis=1)
#orders['student_cat_nlp'] = orders.apply(lambda row: nlp_processing(row, 'Категория ученика:'), axis=1)
#orders.reset_index(drop=True, inplace=True)
#orders.to_feather('orders_nlp.feather')
```

# Обучение моделей 

### Подготовка данных

```python
features = orders_merged[['subject_id', 'price_dif', 'creator_id', 'source_id', 'review_num',
                         'teacher_id', 'rating', 'original_order_id', 'nlp_aim', 'student_cat_nlp',
                        'lesson_cost']]
target = orders_merged['target']
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target,
                                                                            test_size=0.2,
                                                                            stratify=target,
                                                                           random_state=1234)
```

##### Обучение Catboost

```python
categorical = ['nlp_aim', 'student_cat_nlp']

target_encoder = ce.TargetEncoder(cols=categorical)
features_train_encoded = target_encoder.fit_transform(features_train, target_train)
features_test_encoded = target_encoder.transform(features_test)


# Make a scorer from the custom scoring function
scorer = {"AUC": "roc_auc", "f1":'f1'}

param_grid = {
    'depth':  [16]
}
catboost = CatBoostClassifier()
grid_search = GridSearchCV(catboost, 
                           param_grid, 
                           cv=2, 
                           scoring=scorer,
                           n_jobs=-1,
                           refit='AUC',
                           return_train_score=True)
```

```python
grid_search.fit(features_train_encoded, target_train)
```

```python
# Best parameters
print("Best parameters found:", grid_search.best_params_)

# Best AUC score
print("Best AUC score found:", grid_search.best_score_)

# Best estimator
best_cat_model = grid_search.best_estimator_
```

#####  Обучение LGBMClassifier

```python
categorical = ['nlp_aim', 'student_cat_nlp']
target_encoder = ce.TargetEncoder(cols=categorical)
features_train_encoded = target_encoder.fit_transform(features_train, target_train)
features_test_encoded = target_encoder.transform(features_test)

# Now, proceed with training the model

scorer = {"AUC": "roc_auc", "f1":'f1'}

lgbm = LGBMClassifier(class_weight='balanced')
param_grid_LGBM = {'learning_rate':[0.5, 0.1, 0.01],
                'num_leaves':[10, 15, 20],
                'max_depth': [20, 30, 60]
}


lgbm_grid_search = GridSearchCV(lgbm, 
                                param_grid_LGBM, 
                                cv=5, 
                                scoring=scorer, 
                                refit="AUC", 
                                n_jobs=-1) 


```

```python
lgbm_grid_search.fit(features_train_encoded, target_train)
```

```python
# Best parameters
print("Best parameters found:", lgbm_grid_search.best_params_)

# Best AUC score
print("Best AUC score found:", lgbm_grid_search.best_score_)

# Best estimator
best_lgbm_model = lgbm_grid_search.best_estimator_
```

#####  Обучение HistGradientBoostingClassifier

```python
categorical = ['nlp_aim', 'student_cat_nlp']
target_encoder = ce.TargetEncoder(cols=categorical)
features_train_encoded = target_encoder.fit_transform(features_train, target_train)
features_test_encoded = target_encoder.transform(features_test)

scorer = {"AUC": "roc_auc", "f1":'f1'}

model_hb = HistGradientBoostingClassifier(random_state=1234, class_weight='balanced')

param_grid_hb = {
    'learning_rate': [0.5],
    'max_iter': [100, 200],
    'max_depth': [20, 30, 60]
}


hb_grid_search = GridSearchCV(model_hb,
                              param_grid_hb,
                              cv=5,
                              scoring=scorer,
                              refit="AUC",
                              n_jobs=-1) 
```

```python
hb_grid_search.fit(features_train_encoded, target_train)
```

```python
# Best parameters
print("Best parameters found:", hb_grid_search.best_params_)

# Best AUC score
print("Best AUC score found:", hb_grid_search.best_score_)

# Best estimator
best_hb_model = hb_grid_search.best_estimator_



```

#### Сравнение моделей

```python
result = pd.DataFrame(
    [grid_search.best_score_, lgbm_grid_search.best_score_, hb_grid_search.best_score_], 
    index=['Catboost', 'LGBM', 'HistGradientBoostingClassifier'], 
    columns=['AUC']
)

result
```
