```python
data.sort_values(by='date', inplace=True, ascending=False)
```

```python
date_end = pd.Timestamp('2023-9-26')
positive_class=data[data['date_new'] >= date_end]
```

```python
data=data[data['date_new'] < date_end]
```

```python
client_1 =list(positive_class['client_id'].unique())
client_1_set = set(client_1) 
```

```python
data.head()
```

```python
data['time_diff'] =data.groupby('client_id')['date_new'].transform('diff')
```

```python
aggregations = {
          
    'total_price': ['mean', 'max', 'min'],        
     'date_new':['max', 'min'],   
     'quantity_count':'mean',
    'price':['mean', 'max', 'min'],
    'time_diff':['mean', 'min', 'max' ]
}

# Perform the aggregations
aggregated_data = data.groupby('client_id').agg(aggregations)
aggregated_data.reset_index(inplace=True)
```

```python
renamed_columns = {
    ('total_price', 'mean'): 'average_total_price',
    ('total_price', 'max'): 'max_total_price',
    ('total_price', 'min'): 'min_total_price',
    ('date_new', 'max'): 'max_date_new',
    ('date_new', 'min'): 'min_date_new',
    ('quantity_count', 'mean'): 'average_quantity_count',
    ('price', 'mean'): 'average_price',
    ('price', 'max'): 'max_price',
    ('price', 'min'): 'min_price',
    ('time_diff', 'mean'): 'average_time_diff',
    ('time_diff', 'min'): 'min_time_diff',
    ('time_diff', 'max'): 'max_time_diff',
}

aggregated_data.columns =['client_id'] + [renamed_columns[col] for col in aggregated_data.columns[1:]]
```

```python
aggregated_data
```

–î–æ–±–∞–≤–∏–º —Ü–µ–ª–µ–≤—É—é –º–µ—Ç—Ä–∏–∫—É

```python
aggregated_data['target'] = aggregated_data['client_id'].apply(lambda x: 1 if x in client_1_set else 0)
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ

```python
features=aggregated_data.drop(['target', 'client_id'], axis=1)
target=aggregated_data['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, stratify=target)
```

### –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –ø–æ–º–æ—â—å—é k-means

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ k-means

```python
x_train_k=x_train.copy()

x_train_k['average_time_diff_num']=x_train_k['average_time_diff'] / pd.Timedelta(days=1)
x_train_k['min_time_diff_num']=x_train_k['min_time_diff'] / pd.Timedelta(days=1)
x_train_k['max_time_diff_num']=x_train_k['max_time_diff'] / pd.Timedelta(days=1)
x_train_k['days']=(x_train_k['max_date_new']-x_train_k['min_date_new']).dt.days
x_train_k=x_train_k.drop(['average_time_diff', 'min_time_diff', 'max_time_diff', 'max_date_new','min_date_new'], axis=1)
```

```python
x_test_k=x_test.copy()

x_test_k['average_time_diff_num']=x_test_k['average_time_diff'] / pd.Timedelta(days=1)
x_test_k['min_time_diff_num']=x_test_k['min_time_diff'] / pd.Timedelta(days=1)
x_test_k['max_time_diff_num']=x_test_k['max_time_diff'] / pd.Timedelta(days=1)
x_test_k['days']=(x_test_k['max_date_new']-x_test_k['min_date_new']).dt.days
x_test_k=x_test_k.drop(['average_time_diff', 'min_time_diff', 'max_time_diff', 'max_date_new','min_date_new'], axis=1)
```

```python
x_train_k=x_train_k.fillna(0)
x_test_k=x_test_k.fillna(0)
```

```python
# –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è
kmeans = KMeans(n_clusters=2)
kmeans.fit(x_train_k)
cluster_labels = kmeans.labels_

#  –¥–æ–±–∞–≤–ª—è–µ–º –∫–ª–∞—Å—Ç–µ—Ä–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è catboost
x_train['cluster'] = kmeans.predict(x_train_k)
x_test['cluster'] = kmeans.predict(x_test_k)

# –¥–æ–±–∞–≤–ª—è–µ–º –∫–ª–∞—Å—Ç–µ—Ä–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ 
x_train_k['cluster'] = kmeans.predict(x_train_k)
x_test_k['cluster'] = kmeans.predict(x_test_k)

```

```python
x_train['max_date_new']=pd.to_datetime(x_train['max_date_new'])
x_train['min_date_new']=pd.to_datetime(x_train['min_date_new'])
x_test['max_date_new']=pd.to_datetime(x_test['max_date_new'])
x_test['min_date_new']=pd.to_datetime(x_test['min_date_new'])
```

```python
lgbm_grid_search.fit(x_train_k, y_train)
```

```python
# –ü–æ–ª—É—á–µ–Ω–∏–µ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
LGBM_importance_catboost = best_lgbm_model.feature_importances_

# –ü–æ–ª—É—á–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
feature_names = x_test_k.columns

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è —Å –ø–∞—Ä–∞–º–∏ (–Ω–∞–∑–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞, –≤–∞–∂–Ω–æ—Å—Ç—å)
feature_importance_dict = dict(zip(feature_names, LGBM_importance_catboost))

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Å–ª–æ–≤–∞—Ä—è –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –≤–∞–∂–Ω–æ—Å—Ç–∏
sorted_feature_importance = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–∞–∑–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –∏—Ö –≤–∞–∂–Ω–æ—Å—Ç–∏
sorted_features, sorted_importance = zip(*sorted_feature_importance)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_features)), sorted_importance, align='center')
plt.yticks(range(len(sorted_features)), sorted_features)
plt.xlabel('–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∞')
plt.ylabel('–ü—Ä–∏–∑–Ω–∞–∫')
plt.title('–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –º–æ–¥–µ–ª–∏ LGBM')
plt.show()
```
